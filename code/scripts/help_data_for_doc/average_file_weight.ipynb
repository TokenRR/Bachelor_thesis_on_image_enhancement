```python
import os

import matplotlib.pyplot as plt
```

```python
TEST_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\images\\test'
TRAIN_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\images\\train'
VAL_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\images\\val'

RAW_TEST_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\raw_images\\test'
RAW_TRAIN_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\raw_images\\train'
RAW_VAL_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\raw_images\\val'
```

```python
def average_file_size(directory):
    total_size = 0
    file_count = 0
    
    # Перебір всіх файлів у вказаній директорії
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)
        
        # Перевірка чи є шлях файлом
        if os.path.isfile(file_path):
            file_size = os.path.getsize(file_path)
            total_size += file_size
            file_count += 1
    
    # Обчислення середнього розміру файлу
    if file_count == 0:
        return 'No files found in the directory.'
    
    average_size = total_size / file_count
    
    # Переведення у відповідні одиниці
    if average_size < 1024:
        return f'Середня вага файлу: {average_size:.2f} байтів'
    elif average_size < 1024 ** 2:
        return f'Середня вага файлу: {average_size / 1024:.2f} кілобайт'
    else:
        return f'Середня вага файлу: {average_size / (1024 ** 2):.2f} мегабайт'
```

```python
test  = average_file_size(TEST_FOLDER_PATH)
train = average_file_size(TRAIN_FOLDER_PATH)
val   = average_file_size(VAL_FOLDER_PATH)

raw_test  = average_file_size(RAW_TEST_FOLDER_PATH)
raw_train = average_file_size(RAW_TRAIN_FOLDER_PATH)
raw_val   = average_file_size(RAW_VAL_FOLDER_PATH)

print(f'test:  {test}')
print(f'train:  {train}')
print(f'val:  {val}\n')

print(f'raw_test:  {raw_test}')
print(f'raw_train:  {raw_train}')
print(f'raw_val:  {raw_val}')
```

```python
categories = ['test', 'train', 'val']
sizes = [132.35, 126.28, 132.05]

plt.bar(categories, sizes, color='royalblue', width=0.4)
for i, value in enumerate(sizes):
    plt.text(i, value, f"{value:,}", ha='center', va='bottom', fontsize=10)
plt.xlabel('Набори даних')
plt.ylabel('Середній розмір файлу (кілобайти)')
plt.title('Середні розміри файлів у наборах даних')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
