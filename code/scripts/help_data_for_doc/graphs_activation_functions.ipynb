```python
import numpy as np
import matplotlib.pyplot as plt
```

# ReLU
$$
ReLU(x) = \max(0, x)
$$

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.linspace(-10, 10, 400)
y = relu(x)

plt.plot(x, y, label='ReLU', color='royalblue')
plt.xlabel('x')
plt.ylabel('f(x)', rotation=0)
plt.title('ReLU Function')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

# Sigmoid
$$
Sigmoid(x) = \frac{1}{1 + e^{-x}}
$$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 400)
y = sigmoid(x)

plt.plot(x, y, label='Sigmoid', color='royalblue')
plt.xlabel('x')
plt.ylabel('f(x)', rotation=0, labelpad=10)
plt.yticks(np.arange(0, 1.25, 0.25))
plt.title('Sigmoid Function')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

# Tanh
$$
Tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10, 10, 400)
y = tanh(x)

plt.plot(x, y, label='Tanh', color='royalblue')
plt.xlabel('x')
plt.ylabel('f(x)', rotation=0)
plt.title('Tanh Function')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
