```python
import os
from typing import Union

import matplotlib.pyplot as plt
```

```python
TEST_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\images\\test'
TRAIN_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\images\\train'
VAL_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\images\\val'

RAW_TEST_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\raw_images\\test'
RAW_TRAIN_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\raw_images\\train'
RAW_VAL_FOLDER_PATH = 'D:\\KPI\\Bachelor_thesis\\code\\data\\raw_images\\val'
```

```python
def count_files(path: str) -> int:
    """
    Count the total number of files in a directory and its subdirectories.

    Args:
        path (str): The path to the directory.

    Returns:
        int: The total number of files.
    """
    return sum(len(files) for _root, _dirs, files in os.walk(path))
```

```python
files_in_test = count_files(TEST_FOLDER_PATH)
files_in_train = count_files(TRAIN_FOLDER_PATH)
files_in_val = count_files(VAL_FOLDER_PATH)

files_in_raw_test = count_files(RAW_TEST_FOLDER_PATH)
files_in_raw_train = count_files(RAW_TRAIN_FOLDER_PATH)
files_in_raw_val = count_files(RAW_VAL_FOLDER_PATH)

print(f"Total files in test folder  = {files_in_test}")
print(f"Total files in train folder = {files_in_train}")
print(f"Total files in val folder   = {files_in_val}\n")

print(f"Total files in raw_test folder  = {files_in_raw_test}")
print(f"Total files in raw_train folder = {files_in_raw_train}")
print(f"Total files in raw_val folder   = {files_in_raw_val}")
```

```python
categories = ['test', 'train', 'val']
sizes = [files_in_test, files_in_train, files_in_val]

plt.bar(categories, sizes, color='royalblue', width=0.4)
for i, value in enumerate(sizes):
    plt.text(i, value + 1000, f"{value:,}", ha='center', va='bottom', fontsize=10)
plt.yticks(range(0, max(sizes) + 50000, 50000))
plt.xlabel('Набори даних')
plt.ylabel('Кількість зображень')
plt.title('Кількість зображень у відфільтрованих наборах даних')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
categories = ['test', 'train', 'val']
sizes = [files_in_raw_test, files_in_raw_train, files_in_raw_val]

plt.bar(categories, sizes, color='royalblue', width=0.4)
for i, value in enumerate(sizes):
    plt.text(i, value + 1000, f"{value:,}", ha='center', va='bottom', fontsize=10)
plt.yticks(range(0, max(sizes) + 50000, 50000))
plt.xlabel('Набори даних')
plt.ylabel('Кількість зображень')
plt.title('Кількість зображень у початкових наборах даних')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
def get_folder_size(path: str) -> Union[float, str]:
    """
    Calculate the total size of files in a directory and its subdirectories.

    Args:
        path (str): The path to the directory.

    Returns:
        Union[float, str]: The total size of files, formatted as GB, TB, etc.
    """
    total_size = 0
    for root, _dirs, files in os.walk(path):
        for file in files:
            file_path = os.path.join(root, file)
            total_size += os.path.getsize(file_path)
    return format_size(total_size)


def format_size(size: int) -> str:
    """
    Format the size in bytes into a human-readable format.

    Args:
        size (int): The size in bytes.

    Returns:
        str: The formatted size with appropriate unit (e.g., GB, TB).
    """
    for unit in ['bytes', 'KB', 'MB', 'GB', 'TB']:
        if size < 1024.0:
            return f"{size:.2f} {unit}"
        size /= 1024.0
```

```python
weight_test = get_folder_size(TEST_FOLDER_PATH)
weight_train = get_folder_size(TRAIN_FOLDER_PATH)
weight_val = get_folder_size(VAL_FOLDER_PATH)

weight_raw_test = get_folder_size(RAW_TEST_FOLDER_PATH)
weight_raw_train = get_folder_size(RAW_TRAIN_FOLDER_PATH)
weight_raw_val = get_folder_size(RAW_VAL_FOLDER_PATH)

print(f"Total size of test folder  = {weight_test}")
print(f"Total size of train folder = {weight_train}")
print(f"Total size of val folder   = {weight_val}\n")

print(f"Total size of raw_test folder  = {weight_raw_test}")
print(f"Total size of raw_train folder = {weight_raw_train}")
print(f"Total size of raw_val folder   = {weight_raw_val}")
```
